```python
import os
import json
import pandas as pd

def summarize_notebook(filepath):
    try:
        with open(filepath, "r", encoding="utf-8") as f:
            nb = json.load(f)
        cells = nb.get("cells", [])
        md_text = " ".join(
            "".join(cell.get("source", []))
            for cell in cells if cell.get("cell_type") == "markdown"
        )
        code_text = " ".join(
            "".join(cell.get("source", []))
            for cell in cells if cell.get("cell_type") == "code"
        )
        summary = f"{md_text[:500]}...\n\nCode Snippet:\n{code_text[:300]}"
        title = os.path.splitext(os.path.basename(filepath))[0]
        return {"Title": title, "Summary": summary}
    except Exception as e:
        return {"Title": filepath, "Summary": f"Error: {str(e)}"}

# Summarize all .ipynb files in the folder
folder = "notebooks"
notebook_files = [os.path.join(folder, f) for f in os.listdir(folder) if f.endswith(".ipynb")]

summaries = [summarize_notebook(fp) for fp in notebook_files]
df = pd.DataFrame(summaries)
df.head(10)
```

```python
import os
import json

def load_notebook_markdown(notebooks_dir="notebooks"):
    markdown_data = []
    
    for filename in os.listdir(notebooks_dir):
        if filename.endswith(".ipynb"):
            path = os.path.join(notebooks_dir, filename)
            with open(path, "r", encoding="utf-8") as f:
                nb = json.load(f)
            md_cells = [cell["source"] for cell in nb["cells"] if cell["cell_type"] == "markdown"]
            text = "\n".join(["".join(cell) for cell in md_cells])
            markdown_data.append({"filename": filename, "text": text})
    
    return markdown_data

notebook_markdown = load_notebook_markdown()
print(f"Loaded {len(notebook_markdown)} notebooks.")
print(notebook_markdown[0]["filename"])
print(notebook_markdown[0]["text"][:500])
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

# Load model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Convert notebook markdown to embeddings
notebook_texts = [nb["text"] for nb in notebook_markdown]
embeddings = model.encode(notebook_texts, show_progress_bar=True)

# Add to data
for i, nb in enumerate(notebook_markdown):
    nb["embedding"] = embeddings[i]

# Save embeddings as a NumPy array for later use
np.save("notebook_embeddings.npy", embeddings)

# Optionally, also save metadata with filenames
import pandas as pd
metadata_df = pd.DataFrame([{"filename": nb["filename"], "text": nb["text"]} for nb in notebook_markdown])
metadata_df.to_csv("notebook_metadata.csv", index=False)

print("Embeddings generated and saved.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import umap
import pandas as pd

# Load embeddings
embeddings = np.load("notebook_embeddings.npy")

# Load metadata (replace with your actual metadata file if different)
metadata_df = pd.read_csv("notebook_metadata.csv")

# Reduce to 2D using UMAP
reducer = umap.UMAP(random_state=42)
embedding_2d = reducer.fit_transform(embeddings)

# Plot
plt.figure(figsize=(12, 8))
plt.scatter(embedding_2d[:, 0], embedding_2d[:, 1], alpha=0.7)

for i, filename in enumerate(metadata_df['filename']):
    if i % 5 == 0:  # Label every 5th point
        plt.text(embedding_2d[i, 0], embedding_2d[i, 1], str(filename)[:20], fontsize=8)

plt.title("UMAP Projection of Notebook Embeddings")
plt.xlabel("UMAP 1")
plt.ylabel("UMAP 2")
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans

# Cluster into 5 groups (you can change n_clusters)
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
cluster_a = kmeans.fit_predict(embedding_2d)  # <- Store as `cluster_assignments`



# Plot with cluster coloring
plt.figure(figsize=(12, 8))
scatter = plt.scatter(
    embedding_2d[:, 0], embedding_2d[:, 1], 
    c=cluster_a, cmap='tab10', alpha=0.7
)
plt.title("UMAP with KMeans Clusters")
plt.xlabel("UMAP 1")
plt.ylabel("UMAP 2")

# Label a few notebooks
for i, filename in enumerate(metadata_df['filename']):
    if i % 5 == 0:
        plt.text(embedding_2d[i, 0], embedding_2d[i, 1], str(filename)[:20], fontsize=8)

plt.colorbar(scatter, ticks=range(n_clusters), label='Cluster')
plt.show()
```

```python
# Add cluster labels to your metadata
metadata_df['cluster'] = cluster_a

# Print 3 example notebook filenames per cluster
for cluster_id in range(n_clusters):
    print(f"\nðŸŸ¢ Cluster {cluster_id}")
    cluster_files = metadata_df[metadata_df['cluster'] == cluster_id]['filename']
    sample_files = cluster_files.sample(min(3, len(cluster_files)), random_state=42)
    for fname in sample_files:
        print("  -", fname)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Confirm cluster column exists
assert 'cluster' in metadata_df.columns, "No 'cluster' column found in metadata_df"

# Clean text from filenames
metadata_df['text'] = metadata_df['filename'].str.replace(r'[-_]', ' ', regex=True)

# Group filenames per cluster into one large string
cluster_texts = metadata_df.groupby('cluster')['text'].apply(lambda x: ' '.join(x))

# Initialize TF-IDF Vectorizer
vectorizer = TfidfVectorizer(stop_words='english', max_features=20)
X = vectorizer.fit_transform(cluster_texts)

# Extract feature names
feature_names = vectorizer.get_feature_names_out()

# Build summaries dict
summaries = {}

for cluster_id, cluster_vector in zip(cluster_texts.index, X.toarray()):
    top_indices = cluster_vector.argsort()[-5:][::-1]
    top_terms = [feature_names[idx] for idx in top_indices]
    summaries[cluster_id] = {'keywords': top_terms}

# Show results
for cluster_id in summaries:
    print(f"\nCluster {cluster_id} Summary Keywords: {', '.join(summaries[cluster_id]['keywords'])}")
```

```python
import numpy as np

# Step 1: Better function to infer category from filename
def identify_category(filename):
    filename = filename.lower()
    if any(kw in filename for kw in ["variables", "functions", "booleans", "lists", "data-types", "conditionals"]):
        return "Foundations"
    elif any(kw in filename for kw in ["machine-learning", "model", "cross-validation", "categorical", "leakage"]):
        return "Machine Learning"
    elif any(kw in filename for kw in ["charts", "plots", "heatmaps", "distributions", "visual"]):
        return "Visualization"
    elif any(kw in filename for kw in ["neural", "classification", "deep", "underfitting", "overfitting"]):
        return "Deep Learning"
    elif any(kw in filename for kw in ["sql", "missing", "groupby", "file", "reading", "writing", "aggregation"]):
        return "Data Handling"
    else:
        return "Other"

# Step 2: Apply category tagging
metadata_df['category'] = metadata_df['filename'].apply(identify_category)

# Step 3: Generate enhanced summaries and assign category
for cluster_id in summaries.keys():
    cluster_df = metadata_df[metadata_df['cluster'] == cluster_id]
    
    top_category = (
        cluster_df['category'].value_counts().idxmax()
        if not cluster_df.empty else "Unknown"
    )
    summaries[cluster_id]['category'] = top_category

    # Build a more detailed summary
    keywords = summaries[cluster_id].get('keywords', [])
    summary = summaries[cluster_id].get('summary')

    if not summary:
        if top_category == "Foundations":
            summaries[cluster_id]['summary'] = (
                "Introduces programming basics such as variables, conditionals, loops, and data types, "
                "laying the groundwork for more advanced concepts."
            )
        elif top_category == "Machine Learning":
            summaries[cluster_id]['summary'] = (
                "Focuses on essential ML concepts such as model training, validation, handling data leakage, "
                "and encoding strategies. Suitable for beginner-to-intermediate ML learners."
            )
        elif top_category == "Visualization":
            summaries[cluster_id]['summary'] = (
                "Demonstrates how to effectively visualize data using bar plots, heatmaps, and distribution charts, "
                "aiding exploratory data analysis and storytelling with data."
            )
        elif top_category == "Deep Learning":
            summaries[cluster_id]['summary'] = (
                "Explores deep learning foundations including neural network structures, binary classification, "
                "and common pitfalls such as overfitting, with practical examples."
            )
        elif top_category == "Data Handling":
            summaries[cluster_id]['summary'] = (
                "Teaches data preprocessing techniques including handling missing values, file operations, "
                "and data aggregation using SQL and pandas."
            )
        else:
            summaries[cluster_id]['summary'] = (
                f"This cluster includes notebooks covering a variety of topics. Key concepts include: {', '.join(keywords[:5])}."
            )

# Step 4: Display
for cluster_id in sorted(summaries.keys()):
    keywords = summaries[cluster_id].get('keywords', [])
    summary = summaries[cluster_id].get('summary', 'No summary provided.')
    category = summaries[cluster_id].get('category', 'N/A')

    print(f"\nðŸ“Œ Cluster {cluster_id} â€“ Top Keywords: {', '.join(keywords) if keywords else 'N/A'}")
    print(f"ðŸ§  Summary: {summary}")
    print(f"ðŸ“š Model Category: {category}")
    print("ðŸ“ Example Notebooks:")

    example_idxs = metadata_df[metadata_df['cluster'] == cluster_id].index[:5]
    for idx in example_idxs:
        print(f"  - {metadata_df.loc[idx]['filename']}")
```
